```python
import pandas as pd
import numpy as np
```

```python
gold_df = pd.read_csv('https://gold-projections.s3.amazonaws.com/Gold_Prices.csv')
gold_df.head()
```

```python
gold_df = gold_df.dropna().reset_index(drop=True)
gold_df.head()
```

```python
gold_df = gold_df.rename(columns={
    'Name': 'date_recorded',
    'US dollar': 'USA',
    'Euro': 'Euro',
    'Japanese yen': 'Japan',
    'Pound sterling': 'UK',
    'Canadian dollar': 'Canada',
    'Swiss franc': 'Swizerland',
    'Indian rupee': 'India',
    'Chinese renmimbi': 'China',
    'Turkish lira': 'Turkey',
    'Saudi riyal': 'SB',
    'Indonesian rupiah': 'Indonesia',
    'UAE dirham': 'UAE',
    'Thai baht': 'Thailand',
    'Vietnamese dong': 'Vietnam',
    'Egyptian pound': 'Egypt',
    'Korean won': 'SK',
    'Russian ruble': 'Russia',
    'South African rand': 'SA',
    'Australian dollar': 'Australia'
})
```

```python
gold_df.columns
```

```python
gold_df = gold_df.drop(columns=['US dollar.1', 'Euro.1', 'US dollar.2', 'Chinese renmimbi.1', 'Canadian dollar.1'])
gold_df.head()
```

```python
gold_production_df = pd.read_csv('https://gold-projections.s3.amazonaws.com/gold-production.csv')
gold_production_df = gold_production_df.rename(columns={
    'Entity': 'Name',
    'Gold Production (Clio-Infra & USGS)': 'gold_production',
    'Year': 'date_recorded'
})
gold_production_df.head()
```

```python
gold_production_df['Name'] = gold_production_df['Name'].replace({'United States':'USA'})
gold_production_df['Name'] = gold_production_df['Name'].replace({'United Kingdom':'UK'})
gold_production_df['Name'] = gold_production_df['Name'].replace({'Saudi Arabia': 'SB'})
gold_production_df['Name'] = gold_production_df['Name'].replace({'South Korea': 'SK'})
gold_production_df['Name'] = gold_production_df['Name'].replace({'South Africa': 'SA'})
```

```python
inflation_df = pd.read_csv('https://gold-projections.s3.us-east-2.amazonaws.com/inflation_rates.csv')
inflation_df = inflation_df.loc[inflation_df['YEAR'] >= 1978].reset_index(drop=True)
inflation_df = inflation_df.rename(columns={
    'Year': 'date_recorded'
})
inflation_df.head()
```

```python
index_df = pd.read_csv('https://gold-projections.s3.us-east-2.amazonaws.com/us-dollar-index-historical-chart.csv')
index_df = index_df.rename(columns={
    'date': 'date_recorded',
    'value': 'us_dollar_index'
})
index_df.head()
```
